```python
# library imports
import numpy as np
import matplotlib.pyplot as plt
```

# Softplus
The softplus function $\mathrm{softplus}(x)=\mathrm{log}{\{1+\mathrm{exp}{\{x\}}\}}$ is an activation function in the context of artificial neural networks, deep learning models. 

Below, we implement code to evaluate the function, and plot it.

```python
def softplus(x):
    return np.log(1.+np.exp(x))
```

```python
x = np.linspace(-10, 10, 100)
plt.plot(x, softplus(x), label=r'$\mathrm{softplus}(x)=\log{\{1+\exp{\{x\}}\}}$')
plt.legend();
```

The functional form can be found in several contexts, is highly used, and enjoys many benefits. One is the fact that it is differentiable, as opposed to the so-called ReLU. For some scenarios, a disadvantage of the activation function might be the non-negativity of it.

It might be desirable, that for a negative input the activation would be negative, for zero input the activation would be zero, and for positive input the activation would be positive. To accomplish such, while having the shape of the softplus-function, we introduce 'soft'.

# Soft
$\mathrm{soft}(x)=\mathrm{softplus}(x)-\mathrm{softplus}(0)=\mathrm{log}{\{1+\mathrm{exp}{\{x\}}\}}-\mathrm{log}{\{2\}}=\mathrm{log}{\{0.5\cdot(1+\mathrm{exp}{\{x\}})\}}=\mathrm{log}{\{0.5+0.5\mathrm{exp}{\{x\}}\}}$.

Obviously, $\mathrm{softplus}(x)=\mathrm{soft}(x)+\mathrm{log}{\{2\}}$, providing also (further) justification for the naming. 

Below, we implement code to evaluate the soft-function, and plot it along with the softplus.

```python
def soft(x):
    return np.log(0.5+0.5*np.exp(x))
```

```python
x = np.linspace(-10, 10, 100)
plt.plot(x, softplus(x), 'b--', label=r'$\mathrm{softplus}(x)=\mathrm{soft}(x)+\log{\{2\}}$')
plt.plot(x, soft(x), 'k', label=r'$\mathrm{soft}(x)=\log{\{0.5+0.5\exp{\{x\}}\}}$')
plt.legend();
```
